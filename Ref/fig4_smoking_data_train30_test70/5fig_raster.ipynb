```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1)
```

```python
s_train = np.loadtxt('s_train.dat')
print(s_train.shape)
```

```python
s_test = np.loadtxt('s_test.dat')
s_hidden = np.loadtxt('s_hidden_nh14.dat')
s_recover = np.loadtxt('s_recover_EM_nh14.dat')
```

```python
s_recover_PLE = np.loadtxt('s_recover_PLE_nh14.dat')
```

```python
nx,ny = 1,4
nfig = nx*ny
fig, ax = plt.subplots(ny,nx,figsize=(nx*4.5,ny*2.5))

tcut = 80

ax[0].imshow(s_train.T[:,:tcut],cmap='gray',origin='lower')
ax[1].imshow(s_test.T[:,:tcut],cmap='gray',origin='lower')
ax[2].imshow(s_hidden.T[:,:tcut],cmap='gray',origin='lower')
ax[3].imshow(s_recover.T[:,:tcut],cmap='gray',origin='lower')


ax[0].set_title('training set')
ax[1].set_title('original test set')
ax[2].set_title('noisy test set')
ax[3].set_title('recovered test set')

#ax[0].imshow(s_train.T[:,:tcut],origin='lower')
#ax[1].imshow(s_test.T[:,:tcut],origin='lower')
#ax[2].imshow(s_hidden.T[:,:tcut],origin='lower')
#ax[3].imshow(s_recover.T[:,:tcut],origin='lower')

#plt.xlabel('samples')
#plt.ylabel('variable index')
plt.tight_layout(h_pad=1, w_pad=1.5)
#plt.show()

plt.savefig('fig.pdf', format='pdf', dpi=100)
```
