```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

%matplotlib inline
```

```python
np.random.seed(1)
```

```python
s_train = np.loadtxt('s_train.dat')
print(s_train.shape)

h0 = np.loadtxt('h0.dat')
print(h0.shape)

w = np.loadtxt('w2d.dat')
print(w.shape)
```

```python
acc = np.loadtxt('acc.dat')
```

```python
acc_eps = np.loadtxt('acc_eps07_eps098.dat')
```

```python
s_test = np.loadtxt('s_test.dat')
s_hidden = np.loadtxt('s_hidden_nh14.dat')
s_recover = np.loadtxt('s_recover_EM_nh14.dat')
```

```python
#s_recover_PLE = np.loadtxt('s_recover_PLE_nh14.dat')
```

```python
fig = plt.figure(figsize=(10,8))

gs1 = gridspec.GridSpec(3, 2)
gs1.update(left=0.08, right=0.48, wspace=0.3, hspace=0.5)
ax1 = plt.subplot(gs1[0, :])
ax2 = plt.subplot(gs1[1, 0])
ax3 = plt.subplot(gs1[1, 1])
#ax4 = plt.subplot(gs1[2, 0])
ax4 = plt.subplot(gs1[2, :])

gs2 = gridspec.GridSpec(3, 1)
gs2.update(left=0.55, right=0.98, wspace=0.1,hspace=0.5)
ax5 = plt.subplot(gs2[0, 0])
ax6 = plt.subplot(gs2[1, 0])
ax7 = plt.subplot(gs2[2, 0])

#=======================================================
tcut = 101
ax1.set_title('training set')
ax1.imshow(s_train.T[:,:tcut],cmap='gray',origin='lower')

ax2.plot(h0,'ko',markersize=3)
ax2.set_xlabel('variable')
ax2.set_ylabel('local field')

ax3.set_title('interaction matrix')
im = ax3.imshow(w,cmap='rainbow',origin='lower')
plt.colorbar(im,ax=ax3,fraction=0.045, pad=0.05,ticks=[-1.0,0,1.0])
ax3.set_xlabel('variable $i$')
ax3.set_ylabel('variable $j$')

ax4.plot(acc[3:,0],acc[3:,2],'bs--',mfc='none',markersize=5,label='PLE')
ax4.plot(acc[3:,0],acc[3:,3],'ro-',markersize=5,label='EM, $\epsilon$=0.84')
ax4.plot(acc_eps[:,0],acc_eps[:,2],'g^--',mfc='none',markersize=5,label='EM, $\epsilon$=0.70')
ax4.plot(acc_eps[:,0],acc_eps[:,3],'cv--',mfc='none',markersize=5,label='EM, $\epsilon$=0.98')

ax4.set_xlabel('Number of missing variables in each sample')
ax4.set_ylabel('Accuracy')
#ax4.legend()
ax4.legend(labelspacing=0.25,loc='lower left',handletextpad=0.2,borderaxespad=0.2,borderpad=0.1)

#-----------------------------------------------
ax5.set_title('original test set')
ax5.imshow(s_test.T[:,:tcut],cmap='gray',origin='lower')

ax6.set_title('noisy test set')
ax6.imshow(s_hidden.T[:,:tcut],cmap='gray',origin='lower')

ax7.set_title('recovered test set')
ax7.imshow(s_recover.T[:,:tcut],cmap='gray',origin='lower')

ax1.set_xlabel('sample')
ax1.set_ylabel('variable')

ax5.set_xlabel('sample')
ax5.set_ylabel('variable')

ax6.set_xlabel('sample')
ax6.set_ylabel('variable')

ax7.set_xlabel('sample')
ax7.set_ylabel('variable')

ax4.set_xticks([8,10,12,14,16])
ax2.set_xticks([0,10,20,30])
ax2.set_yticks([-1,0,1])
ax3.set_xticks([0,10,20,30])
ax3.set_yticks([0,10,20,30])

# manipulate
vals = ax4.get_yticks()
ax4.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
   
ax1.text(-0.05,1,'(a)',transform=ax1.transAxes,va='top',ha='right',fontsize=13)
ax2.text(-0.12,1,'(b)',transform=ax2.transAxes,va='top',ha='right',fontsize=13)
ax3.text(-0.1,1,'(c)',transform=ax3.transAxes,va='top',ha='right',fontsize=13)
ax4.text(-0.09,1,'(g)',transform=ax4.transAxes,va='top',ha='right',fontsize=13)
ax5.text(-0.05,1,'(d)',transform=ax5.transAxes,va='top',ha='right',fontsize=13)
ax6.text(-0.05,1,'(e)',transform=ax6.transAxes,va='top',ha='right',fontsize=13)
ax7.text(-0.05,1,'(f)',transform=ax7.transAxes,va='top',ha='right',fontsize=13)

plt.tight_layout(h_pad=0.5, w_pad=0.5)
plt.savefig('fig4_eps.pdf', format='pdf', dpi=100)
```
