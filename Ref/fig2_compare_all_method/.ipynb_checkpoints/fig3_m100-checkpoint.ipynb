```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as nplin
import itertools
from sklearn.linear_model import LinearRegression

from matplotlib.ticker import MultipleLocator,FormatStrFormatter,AutoMinorLocator,ScalarFormatter
```

```python
np.random.seed(0)
```

```python
label = ['(a)','(b)','(c)','(d)']

nx,ny = 2,2
nfig = nx*ny
fig, ax = plt.subplots(ny,nx,figsize=(nx*3.5,ny*2.9))

for i,g in enumerate([0.5,1.0]):
    w = np.loadtxt('m100/w_100_%s_20000.dat'%g)
    ax[0,i].plot([-1,1],[-1,1],'k--')
    ax[0,i].plot(w[0],w[1],'m^',marker='^',mfc='none',markersize=4,label='HF')
    #ax[0,i].plot(w[0],w[4],'go',marker='o',mfc='none',markersize=4,label='RE')
    ax[0,i].plot(w[0],w[5],'ro',marker='o',markersize=4,label='EM')
    
    #ax[0,i].set_xlim([-0.8,0.8])
    #ax[0,i].set_ylim([-0.8,0.8])
    ax[0,i].set_xlabel('Actual interactions')
    ax[0,i].set_ylabel('Inferred interactions')

ax[0,0].set_xlim([-0.3,0.3])
ax[0,0].set_ylim([-0.3,0.3])
ax[0,1].set_xlim([-0.6,0.6])
ax[0,1].set_ylim([-0.6,0.6])

ax[0,0].set_xticks([-0.3,0,0.3])
ax[0,0].set_yticks([-0.3,0,0.3])
ax[0,1].set_xticks([-0.6,-0.3,0,0.3,0.6])
ax[0,1].set_yticks([-0.6,-0.3,0,0.3,0.6])

ax[0,0].legend()
ax[0,0].set_title('$M=100$, weak interaction')
ax[0,1].set_title('$M=100$, strong interaction')

# fig c)
t1 = np.loadtxt('fig3d_time/run_time_MLE.txt')
t2 = np.loadtxt('fig3d_time/run_time_pseudo.txt')
t3 = np.loadtxt('fig3d_time/run_time_eps_random.txt')

ax[1,0].semilogy(t1[0],t1[1],'kv--',mfc='none',markersize=5,label='MLE')
ax[1,0].semilogy(t2[0,:-4],t2[1,:-4],'bs--',mfc='none',markersize=5,label='PLE')
#ax[1,0].semilogy(t3[0],t3[1],'go--',mfc='none',markersize=4,label='random $\epsilon$')
ax[1,0].semilogy(t3[0],t3[1]*9,'ro-',markersize=5,label='EM')
ax[1,0].set_xlabel('System size')
ax[1,0].set_ylabel('Computational time [s]')
ax[1,0].legend()
ax[1,0].set_ylim([1,2000])


ax[1,1].set_xlabel('Actual probability')
ax[1,1].set_ylabel('Inferred probability')

xlabel = np.full(nfig,-0.2)
ylabel = np.full(nfig,1.)
k = 0
for i in range(ny):
    for j in range(nx):        
        ax[i,j].text(xlabel[k],ylabel[k],label[k],transform=ax[i,j].transAxes,va='top',ha='right',fontsize=13)
        k += 1

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.savefig('fig3.pdf', format='pdf', dpi=100)
```
