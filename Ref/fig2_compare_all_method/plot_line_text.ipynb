```python
nx,ny = 3,6
nfig = nx*ny
fig, ax = plt.subplots(ny,nx,figsize=(nx*3.5,ny*2.8))




label = ['(a)','(b)','(c)','(d)','(e)','(f)','(g)','(h)','(i)','(j)','(k)','(l)','(m)','(n)',
        '(o)','(p)','(q)','(r)','(s)','(t)','(u)','(v)','(w)','(x)','(y)','(z)']

xlabel = np.full(nfig,-0.2)
ylabel = np.full(nfig,1.)
k = 0
for i in range(ny):
    for j in range(nx):        
        ax[i,j].text(xlabel[k],ylabel[k],label[k],transform=ax[i,j].transAxes,va='top',ha='right',fontsize=13)
        k += 1

ax[0,0].text(-0.5,0,'$M=20$',transform=ax[0,0].transAxes,va='top',ha='right',fontsize=13,rotation=90)
ax[2,0].text(-0.5,0,'$M=40$',transform=ax[2,0].transAxes,va='center',ha='right',fontsize=13,rotation=90)
ax[4,0].text(-0.5,0,'$M=100$',transform=ax[4,0].transAxes,va='center',ha='right',fontsize=13,rotation=90)

#ax[0,0].text(-0.4,0,'weak interaction',transform=ax[0,0].transAxes,va='top',ha='right',fontsize=13,rotation=90)


plt.tight_layout(h_pad=0.7, w_pad=1.5)
#plt.savefig('fig2.pdf', format='pdf', dpi=100)
```
