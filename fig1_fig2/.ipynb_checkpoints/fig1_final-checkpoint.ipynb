```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as nplin
import itertools
```

```python
np.random.seed(0)
```

```python
MSE = np.loadtxt('MSE_l5k.txt')
KL = np.loadtxt('KL_l5k.txt')
E_av = np.loadtxt('E_av_l5k.txt')
```

```python
MSE2 = np.loadtxt('MSE_l10k.txt')
KL2 = np.loadtxt('KL_l10k.txt')
E_av2 = np.loadtxt('E_av_l10k.txt')
```

```python
E_av = -E_av
#E_av2 = -E_av2
```

```python
eps_list = np.linspace(0,1.0,11)
n_eps = len(eps_list)

MSE_eps = np.zeros(n_eps)
KL_eps = np.zeros(n_eps)
E_eps = np.zeros(n_eps)
for i in range(n_eps):
    it = np.argmin(KL[i,:])
    MSE_eps[i] = MSE[i,it]
    KL_eps[i] = KL[i,it]
    E_eps[i] = E_av[i,:].min()
```

```python
MSE2_eps = np.zeros(n_eps)
KL2_eps = np.zeros(n_eps)
E2_eps = np.zeros(n_eps)
for i in range(n_eps):
    it = np.argmin(KL2[i,:])
    MSE2_eps[i] = MSE2[i,it]
    KL2_eps[i] = KL2[i,it]
    E2_eps[i] = E_av2[i,:].min()
```

```python
#label = ['(a)','(b)','(c)','(d)','(e)','(f)','(g)','(h)','(i)','(j)','k','l']
label = ['(a)','(e)','(b)','(f)','(c)','(g)','(d)','(h)']

ieps_show = [1,5,8]

nx,ny = 2,4
nfig = nx*ny
fig, ax = plt.subplots(ny,nx,figsize=(nx*3.5,ny*3))

ax[0,0].plot(MSE[1],'k--',label='$\epsilon=0.1$')
ax[0,0].plot(MSE[5],'r-',label='$\epsilon=0.5$')
ax[0,0].plot(MSE[8],'b--',label='$\epsilon=0.8$')

ax[1,0].plot(KL[1],'k--',label='eps=0.1')
ax[1,0].plot(KL[5],'r-',label='eps=0.5')
ax[1,0].plot(KL[8],'b--',label='eps=0.8')

ax[2,0].plot(E_av[1],'k--',label='eps=0.1')
ax[2,0].plot(E_av[5],'r-',label='eps=0.5')
ax[2,0].plot(E_av[8],'b--',label='eps=0.8')

ax[3,0].plot(KL[1],MSE[1],'ko',marker='s',mfc='none',markersize=5,label='$\epsilon=0.1$')
ax[3,0].plot(KL[5],MSE[5],'ro',marker='o',markersize=5,label='$\epsilon=0.5$')
ax[3,0].plot(KL[8],MSE[8],'bo',marker='^',markersize=5,mfc='none',label='$\epsilon=0.8$')    

####
ax[0,1].plot(eps_list[:],MSE_eps[:],'ko-',markersize=5,label='$L=5,000$')
ax[0,1].plot(eps_list[:],MSE2_eps[:],'bs--',mfc='none',markersize=5,label='$L=10,000$')

ax[1,1].plot(eps_list[:],KL_eps[:],'ko-',markersize=5,label='$L=5000$')
ax[1,1].plot(eps_list[:],KL2_eps[:],'bs--',mfc='none',markersize=5,label='$L=10000$')

ax[2,1].plot(eps_list[:],E_eps[:],'ko-',markersize=5,label='$L=5000$')
ax[2,1].plot(eps_list[:],E2_eps[:],'bs--',mfc='none',markersize=5,label='$L=10000$')

ax[3,1].plot(E_eps,MSE_eps,'ko',markersize=5,label='$L=5,000$')
ax[3,1].plot(E2_eps,MSE2_eps,'bs',markersize=5,mfc='none',label='$L=10,000$')

#============================
ax[0,0].set_ylim([0,0.01])
ax[1,0].set_ylim([4,9])

ax[0,0].set_ylabel('Mean squared error')
ax[0,0].set_xlabel('Iterations')

ax[1,0].set_ylabel('Kullback-Leibler divergence')
ax[1,0].set_xlabel('Iterations')

ax[2,0].set_ylabel('Energy')
ax[2,0].set_xlabel('Iterations')

ax[3,0].set_ylabel('Mean squared error')
ax[3,0].set_xlabel('Kullback-Leibler divergence')

ax[0,1].set_ylabel('Mean squared error')
ax[0,1].set_xlabel(r'$\epsilon$')

ax[1,1].set_ylabel('Kullback-Leibler divergence')
ax[1,1].set_xlabel(r'$\epsilon$')

ax[2,1].set_ylabel('Energy')
ax[2,1].set_xlabel(r'$\epsilon$')

ax[3,1].set_ylabel('Mean squared error')
ax[3,1].set_xlabel('Energy')

ax[0,0].legend()
ax[0,1].legend()
ax[3,0].legend()
ax[3,1].legend()

ax[0,0].set_yticks([0,0.005,0.01])
ax[0,1].set_yticks([0,0.005,0.01])
ax[0,1].set_ylim([0,0.01])

ax[3,0].set_yticks([0,0.05,0.1])
ax[3,1].set_yticks([0,0.005,0.01])

ax[3,0].set_xticks([5.,10.,15.])

xlabel = np.full(nfig,-0.2)
ylabel = np.full(nfig,1.)
k = 0
for i in range(ny):
    for j in range(nx):        
        ax[i,j].text(xlabel[k],ylabel[k],label[k],transform=ax[i,j].transAxes,va='top',ha='right',fontsize=13)
        k += 1
        
plt.tight_layout(h_pad=1, w_pad=1.5)
plt.savefig('fig1_final.pdf', format='pdf', dpi=100)
```
