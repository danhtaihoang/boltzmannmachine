```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.loadtxt('Z_g1.txt').T
```

```python
z0 = 19.06457473559291
```

```python
plt.plot(1/np.log(z[0]),z[2],'ko-',label='unique')
plt.plot(1/np.log(z[0]),z[3],'bs-',label='observed')
plt.plot(1/np.log(z[0]),z[4],'r^-',label='unique-optimal-eps')
plt.plot([0,1/np.log(z[0].min())],[z0,z0],'r--',label='exact')
plt.xlim([1/np.log(2**20),1/np.log(z[0].min())])
plt.xlabel('number of observed sequences')
plt.legend()
```

```python
plt.plot(1/np.log(z[1]),z[2],'ko-',label='unique')
plt.plot(1/np.log(z[1]),z[3],'bs-',label='observed')
plt.plot(1/np.log(z[1]),z[4],'r^-',label='unique-optimal-eps')
plt.plot([0,1/np.log(z[1].min())],[z0,z0],'r--',label='exact')
plt.xlim([1/np.log(2**20),1/np.log(z[1].min())])
plt.xlabel('number of unique sequences')
plt.legend()
```
